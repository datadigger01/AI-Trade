```python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
import matplotlib.pyplot as plt
```

```python
#symbol = 'BTC-USD'
#period = '1mo'
#price_data = yf.download(symbol, period=period)

symbol = 'KC=F'
period = '30y'
price_data = yf.download(symbol, period=period)
price_data
```

```python
price_data.info()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(price_data.index, price_data[('Close', 'KC=F')], label='Coffee Price')
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
