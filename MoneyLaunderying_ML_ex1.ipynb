## 원데이터에서부터 새로운 Feature 파생
   - Time과 Date 컬럼으로 부터 송금시간, 송금일자 추출

```python
#df 데이터셋에서 time 컬럼 시간을 추출하고 싶고, date 컬럼에서는 일자를 추출하려면 어떻게 해야 하는지 알려주세요.
df['Extracted_Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
df['Extracted_Date'] = pd.to_datetime(df['Date']).dt.date

print("Time and Date successfully extracted into 'Extracted_Time' and 'Extracted_Date' columns.")
print(df[['Time', 'Extracted_Time', 'Date', 'Extracted_Date']].head())
```

```python
# 내가 원하는 것은 예를 들어 time 컬럼값에 10:35:19가 들어 있으면 10을 추출하고, date 컬럼값에 예를 들어 2022-10-07값이 들어 있다면 7일자값만 추출하고 싶어요.

df['Hour_Only'] = df['Extracted_Time'].apply(lambda x: x.hour)
df['Day_Only'] = df['Extracted_Date'].apply(lambda x: x.day)

print("Hour and Day successfully extracted into 'Hour_Only' and 'Day_Only' columns.")
print(df[['Time', 'Hour_Only', 'Date', 'Day_Only']].head())
```

 - 송금국가와 수신국가 일치여부

```python
# df 데이터셋에서 Sender_bank_location 컬럼값과 Receiver_bank_location 컬럼값이 일치하면 0, 일치하지 않으면 1로 설정하여 Send_Rec_YN 컬럼명으로 새로 만들어주세요.

df['Send_Rec_YN'] = (df['Sender_bank_location'] != df['Receiver_bank_location']).astype(int)

print(df[['Sender_bank_location', 'Receiver_bank_location', 'Send_Rec_YN']].head())
```

 - 송금통화와 수금통화 일치여부

```python
df['SendCur_RecCur_YN'] = (df['Payment_currency'] != df['Received_currency']).astype(int)

print(df[['Payment_currency', 'Received_currency', 'SendCur_RecCur_YN']].head())
```

### 데이터탐색(EDA) Feature와 Label간의 관계를 시각적으로 파악

```python
#Prompt: 'Payment_currency', 'Received_currency', 'Sender_bank_location', 'Receiver_bank_location', 'Payment_type', 'Laundering_type', 'Hour_Only','Day_Only','Send_Rec_YN','SendCur_RecCur_YN' 컬럼에 대해서 왼쪽축 기준으로는 정렬된 컬럼값 기준의 빈도수를
# 오른쪽축기준으로는 'Is_laundering' 값에 대한 발생비율이 최대값이 0.05으로 고정된 그래프로 그리되, 3열로 보여지게 하는 코드를 제공해주세요.
# 그리고 막대그래프를  연한회색으로하고 선그래프는 파란색으로 셋팅해주세요
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns_to_plot = ['Payment_currency', 'Received_currency', 'Sender_bank_location', 'Receiver_bank_location',
                   'Payment_type', 'Laundering_type', 'Hour_Only', 'Day_Only', 'Send_Rec_YN', 'SendCur_RecCur_YN']

num_cols = 2
num_rows = (len(columns_to_plot) + num_cols - 1) // num_cols

plt.figure(figsize=(num_cols * 6, num_rows * 5))

for i, col in enumerate(columns_to_plot):
    plt.subplot(num_rows, num_cols, i + 1)
    ax1 = sns.countplot(x=col, data=df, order=df[col].value_counts().index, palette=['#D3D3D3'] * len(df[col].unique()), hue=col, legend=False)
    ax1.set_xlabel(col)
    ax1.set_ylabel('Frequency', color='gray')
    ax1.tick_params(axis='y', labelcolor='gray')
    ax1.tick_params(axis='x', rotation=45) # Rotate x-axis labels
    plt.setp(ax1.get_xticklabels(), ha='right') # Set horizontal alignment for rotated labels
    ax1.set_title(f'Frequency & Laundering Ratio by {col}')

    ax2 = ax1.twinx()

    # Extract labels based on column type for correct .loc indexing
    if col in ['Hour_Only', 'Day_Only', 'Send_Rec_YN', 'SendCur_RecCur_YN']: # Added 'Send_Rec_YN', 'SendCur_RecCur_YN'
        labels = [int(label.get_text()) for label in ax1.get_xticklabels()]
    else:
        labels = [label.get_text() for label in ax1.get_xticklabels()]

    ratio = df.groupby(col)['Is_laundering'].mean().loc[labels]

    sns.lineplot(x=ax1.get_xticks(), y=ratio, ax=ax2, color='blue', marker='o')
    ax2.set_ylabel('Laundering Ratio', color='blue')
    ax2.tick_params(axis='y', labelcolor='blue')
    ax2.set_ylim(0, 0.01) # Fixed max at 0.01

plt.tight_layout()
plt.show()
```
